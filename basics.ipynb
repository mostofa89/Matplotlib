Graph Title and Axis Labels

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]


plt.plot(x, y)
plt.title("Simple Plot")
plt.show()
```

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]


plt.plot(x, y)

plt.title("Simple Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

Graph Markers and Line Styles

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]
```

```python
plt.plot(x, y, marker='o', linestyle='--', color='r')

plt.title("Simple Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
plt.plot(x, y, marker='+', linestyle='--', color='r')

plt.title("Simple Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
plt.plot(x, y, marker='>', linestyle='--', color='r')

plt.title("Simple Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
plt.plot(x, y, marker='o', linestyle='--', color='r', markeredgecolor = "g")

plt.title("Simple Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
plt.plot(x, y, 'r^--')

plt.title("Simple Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
plt.plot(x, y, 'gx--')

plt.title("Simple Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
plt.plot(x, y, 'b*--')

plt.title("Simple Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
plt.plot(x, y, 'yo--')

plt.title("Simple Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

Line Width and Marker Labels

```python
plt.plot(x, y, linestyle='--', color='r', markersize = 10, linewidth = 12)

plt.title("Simple Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
plt.plot(x, y, linestyle='--', color='r', label = '2x')

plt.title("Simple Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.legend()
plt.show()
```

Graph Ticks and Grid

```python
plt.plot(x, y, linestyle='--', color='r', label = '2x')

plt.title("Simple Plot")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8, 8.5])

plt.legend()
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])

plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 1, 2, 3, 4])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])

plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 1, 2, 3, 4])
plt.grid()

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 2, 3], linewidth = 3, markersize = 10, marker = 'o', markeredgecolor = 'r', markerfacecolor = 'g')

plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 1, 2, 3, 4])
plt.grid()

plt.show()
```

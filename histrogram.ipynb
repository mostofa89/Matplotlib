```python
ticks = [i for i in range(40, 101, 10)]

plt.hist(fifa.Overall, bins=ticks)
plt.xticks(ticks)
plt.xlabel('sets of Overall Ratings')
plt.ylabel('Number of Players')
plt.title('Distribution of Player Ratings in FIFA 2018')

plt.show()
```

```python
ticks = [i for i in range(40, 101, 10)]

plt.hist(fifa.Overall, bins=ticks, color='green', edgecolor='black', alpha=0.7)
plt.xticks(ticks)
plt.xlabel('sets of Overall Ratings')
plt.ylabel('Number of Players')
plt.title('Distribution of Player Ratings in FIFA 2018')

plt.show()
```

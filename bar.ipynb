```python
label = ["A", "B", "C"]
values = [2, 4, 2]

bars = plt.bar(label, values)
plt.title("Bar Graph")

bars[0].set_hatch("/")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
label = ["A", "B", "C"]
values = [2, 4, 2]

bars = plt.bar(label, values)
plt.title("Bar Graph")

bars[0].set_hatch("/")
bars[1].set_hatch("*")
bars[2].set_hatch("o")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
label = ["A", "B", "C", "D", "E", "F"]
values = [2, 4, 2, 5, 6, 7]

bars = plt.bar(label, values)
plt.title("Bar Graph")
patterns = ["/", "*", "o", "\\", "-", "+"]
for bar, pattern in zip(bars, patterns):
    bar.set_hatch(pattern)


plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
label = ["A", "B", "C", "D", "E", "F"]
values = [2, 4, 2, 5, 6, 7]

bars = plt.bar(label, values)
plt.title("Bar Graph")
patterns = ["/", "*", "o", "\\"]
for bar in bars:
    hatch = patterns[bars.index(bar) % len(patterns)]
    bar.set_hatch(hatch)



plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
label = ["A", "B", "C", "D", "E", "F"]
values = [2, 4, 2, 5, 6, 7]

bars = plt.bar(label, values)
plt.title("Bar Graph")
patterns = ["/", "*", "o", "\\"]
for bar in bars:
    hatch = patterns.pop(0)
    bar.set_hatch(hatch)
    patterns.append(hatch)



plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
label = ["A", "B", "C", "D", "E", "F"]
values = [2, 4, 2, 5, 6, 7]

bars = plt.bar(label, values)
plt.title("Bar Graph")
for bar in bars:
    bar.set_color('cyan')



plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
label = ["A", "B", "C"]
values = [2, 4, 2]

bars = plt.bar(label, values)
plt.title("Bar Graph")

bars[0].set_color("Green")
bars[1].set_color("Red")
bars[2].set_color("Blue")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

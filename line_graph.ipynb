Line Graph

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv("csvFiles\gas_prices.csv")

print(gas.to_string())  # Print the entire DataFrame
```

```python
plt.plot(gas.Year, gas.USA)

plt.show()
```

```python
plt.plot(gas.Year, gas.USA, marker='o', linestyle='--', color='g', label='USA Gas Prices')

plt.show()
```

```python
plt.plot(gas.Year, gas.Japan)

plt.show()
```

```python
plt.plot(gas.Year, gas.USA, label='USA')
plt.plot(gas.Year, gas.Japan, label='Japan')
plt.plot(gas.Year, gas.Canada, label='Canada')
plt.plot(gas.Year, gas.France, label='France')

plt.xlabel("Year")
plt.ylabel("Gas Price (USD)")
plt.title("Gas Prices Over Time")

plt.legend()
plt.show()
```

```python
plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
plt.plot(gas.Year, gas.Japan, 'y.-', label='Japan')
plt.plot(gas.Year, gas.Canada, 'g.-', label='Canada')
plt.plot(gas.Year, gas.France, 'r.-', label='France')

print(gas.Year[: 3])

plt.xlabel("Year")
plt.ylabel("Gas Price (USD)")
plt.title("Gas Prices Over Time")
plt.xticks(gas.Year[::3])  # Set x-ticks to every 3rd year

plt.legend()
plt.show()
```

```python
plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
plt.plot(gas.Year, gas.Japan, 'y.-', label='Japan')
plt.plot(gas.Year, gas.Canada, 'g.-', label='Canada')
plt.plot(gas.Year, gas.France, 'r.-', label='France')

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker = '.', label = country)


plt.xlabel("Year")
plt.ylabel("Gas Price (USD)")
plt.title("Gas Prices Over Time")


plt.legend()
plt.show()
```

```python
plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
plt.plot(gas.Year, gas.Japan, 'y.-', label='Japan')
plt.plot(gas.Year, gas.Canada, 'g.-', label='Canada')
plt.plot(gas.Year, gas.France, 'r.-', label='France')

print(gas.Year[: 3])

plt.xlabel("Year")
plt.ylabel("Gas Price (USD)")
plt.title("Gas Prices Over Time")
plt.xticks(gas.Year[::3].to_list() + [2010])  # Set x-ticks to every 3rd year

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5), dpi=300)
plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
plt.plot(gas.Year, gas.Japan, 'y.-', label='Japan')
plt.plot(gas.Year, gas.Canada, 'g.-', label='Canada')
plt.plot(gas.Year, gas.France, 'r.-', label='France')
print(gas.Year[: 3])

plt.xlabel("Year")
plt.ylabel("Gas Price (USD)")
plt.title("Gas Prices Over Time")
plt.xticks(gas.Year[::3].to_list() + [2010])  # Set x-ticks to every 3rd year

plt.savefig("gas_prices.png", dpi=300)
plt.legend()
plt.show()
```

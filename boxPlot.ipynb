Box Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
fifa = pd.read_csv('csvFiles/fifa_data.csv')
```

```python
barca_players = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid_players = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
```

```python
plt.boxplot([barca_players, madrid_players])
```

```python
plt.boxplot([barca_players, madrid_players])
plt.show()
```

```python
plt.boxplot([barca_players, madrid_players])
plt.xticks([1, 2], ['FC Barcelona', 'Real Madrid'])
plt.ylabel('Overall Rating')
plt.title('Overall Ratings of FC Barcelona and Real Madrid Players')
plt.show()
```

```python
labels = ['FC Barcelona', 'Real Madrid']

plt.boxplot([barca_players, madrid_players], labels=labels)
plt.title('Overall Ratings of FC Barcelona and Real Madrid Players')

plt.show()
```

```python
manCity_players = fifa.loc[fifa.Club == 'Manchester City']['Overall']
manUtd_players = fifa.loc[fifa.Club == 'Manchester United']['Overall']
```

```python
label = ['FC Baecelona', 'Real Marid', 'Manchester City', 'Manchester United']

plt.boxplot([barca_players, madrid_players, manCity_players, manUtd_players], labels=label)
plt.title('Overall Ratings of Top 4 Clubs Players')
plt.ylabel('Overall Rating')


plt.show()
```

```python
label = ['FC Baecelona', 'Real Marid', 'Manchester City', 'Manchester United']
plt.figure(figsize=(7, 8))

plt.boxplot([barca_players, madrid_players, manCity_players, manUtd_players], labels=label)
plt.title('Overall Ratings of Top 4 Clubs Players')
plt.ylabel('Overall Rating')


plt.show()
```

```python
label = ['FC Baecelona', 'Real Marid', 'Manchester City', 'Manchester United']
plt.figure(figsize=(7, 8))

boxes = plt.boxplot([barca_players, madrid_players, manCity_players, manUtd_players], labels=label)
plt.title('Overall Ratings of Top 4 Clubs Players')
plt.ylabel('Overall Rating')

for box in boxes['boxes']:
    box.set(color='#4287f5', linewidth=2)


plt.show()
```

```python
label = ['FC Baecelona', 'Real Marid', 'Manchester City', 'Manchester United']
plt.figure(figsize=(7, 8))

boxes = plt.boxplot([barca_players, madrid_players, manCity_players, manUtd_players], labels=label, patch_artist=True, medianprops={'linewidth': 2, 'color': 'orange'})
plt.title('Overall Ratings of Top 4 Clubs Players')
plt.ylabel('Overall Rating')

for box in boxes['boxes']:
    box.set(color="#353536", linewidth=2)
    box.set(facecolor = "#538c5f")
    box.set(hatch = '*')


plt.show()
```
